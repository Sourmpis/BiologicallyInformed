```python
# parameter_recovery : GoNoGo_nofb, GoNoGo_withfb1, GoNoGo_withfb2 # fb2 is the differential
opt_default = SimpleNamespace(**default)
for seed in [0,1, 2,3,4]:
    for l1 in [0.02]:
        opt_student = copy.deepcopy(opt_default)
        opt_student.batch_size = 150
        opt_student.l1_decay_across = l1
        opt_student.seed = seed
        
        # no spike
        opt_student.spike_function = "sigmoid"
        # non-local inhibition 
        opt_student.restrict_inter_area_inh = False
        # # no tm loss
        # opt_student.loss_trial_wise = 0
        # No Dale's law
        opt_student.flag_ei = False

        string = "_sigmaRNN"

        opt_student = vars(opt_student)
        opt_student["device"] = "cuda"
        
        if not os.path.exists(f"configs/grid_Vahid/l1across{int(l1*10000)}_seed{seed}{string}"):
            os.makedirs(f"configs/grid_Vahid/l1across{int(l1*10000)}_seed{seed}{string}")
        json.dump(opt_student, open(f"configs/grid_Vahid/l1across{int(l1*10000)}_seed{seed}{string}/opt.json", "w"))
# opt_student["loss_trial_wise"] = 0
# if not os.path.exists(f"configs/grid_{name}/l1across{int(l1*10000)}_notm"):
#     os.makedirs(f"configs/grid_{name}/l1across{int(l1*10000)}_notm")
# json.dump(opt_student, open(f"configs/grid_{name}/l1across{int(l1*10000)}_notm/opt.json", "w"))
```
